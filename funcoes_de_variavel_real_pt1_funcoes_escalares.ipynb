 ## 0.2. Funções e suas inversas

 **Definição**

Uma função é uma relação de um conjunto ${\textstyle A}$ com um conjunto ${\textstyle B}$. Denotamos uma função por

\begin{align}
  f : A ⇒ B \\
  y = f(x)
    \end{align}
  
onde ${\textstyle f}$ é o nome da função, ${\textstyle A}$ é chamado de domínio, ${\textstyle B}$ é chamado de contradomínio e ${\textstyle y=f(x)}$ expressa a lei de formação (relação) dos elementos, ${\textstyle x\in A}$ com os elementos ${\textstyle y\in B}$. Considerando o conjunto de pares ordenados ${\textstyle (x,y)}$ de ${\textstyle A}$ x ${\textstyle B}$, teremos uma relação entre os elementos de ${\textstyle A}$ e de ${\textstyle B}$ ou, simplesmente, relação binária de ${\textstyle A}$ em ${\textstyle B}$. Para cada elemento do domínio, existirá um único correspondente no contradomínio, esse correspondente é conhecido como imagem.

É possível classificar uma função de acordo com as características da relação estabelecida entre domínio e contradomínio, são elas:

1. Injetiva
2. Bijetiva
3. Sobrejetiva

**Inversão de funções**

Em matemática, a função inversa de uma função  ${\displaystyle f:X\rightarrow Y}$ é, quando existe, a função ${\displaystyle f^{-1}:Y\rightarrow X}$ tal que ${\displaystyle f\circ f^{-1}=\mathrm {id} _{X}}$ e $f ^{− 1} ∘ f = i d_{Y}$ (id = função identidade). Ou seja, o que era domínio na função original (o conjunto ${\displaystyle X}$ neste caso, ilustrado na figura abaixo) vira imagem na função inversa, e o que era imagem na função original ( ${\displaystyle Y}$, neste caso - ilustrado na figura abaixo) vira domínio.

Uma função só admite inversa à esquerda, se, e somente se, a função for Injetora, e à direita se a função for Sobrejetora. Em outras palavras, quando uma função admite uma inversa, o domínio da função $f$ será o contradomínio da função $f{-1}$

Se tratamos de uma função definida nos reais, tal que $f:R ⇒ R$ é uma função bijetora onde $y = f(x)$, então para obtermos a sua inversa podemos reescrever a função da forma $x = g(y)$ tal que x dependa de y. Com isso temos que esta nova função $g(y) = f^{-1}$

Graficamente uma função e sua inversa podem ser relacionadas por uma operação de reflexão em torno da reta $y = x$

## 1.3. Definindo as funções

As funções em python podem ser definidas de várias maneiras, nesse documento trataremos da definição utilizando "def" e "lambda/anônima"

````python
# Definindo na forma anônima
a = lambda x: x**2 + 2*x + 3  # polinomial do grau 2
b = lambda x: 3*x**3 - 2*x**2 + 4*x + 5  # polinomial do grau 3

# Definindo utilizando "def"
def c(x):
    return 10 * a(x) / b(x)  # função racional
````

## 1.4. Lista com as principais funções com numpy

Utilizando a biblioteca numpy "np." é possível definir diversas funções de maneira muito simples. Além disso é possível manipular o domínio da função com o comando "np.where(condition, value_if_true, value_if_false!)". Segue uma lista com as principais funções:


````python
# Funções Polinomial, racional, raiz e modular
a = lambda x: x**2 + 2*x + 3  # polinomial do grau 2
b = lambda x: 3*x**3 - 2*x**2 + 4*x + 5  # polinomial do grau 3
c = lambda x: 100 * a(x) / b(x)  # função racional
d = lambda x: np.where(x > 0, 10 * np.sqrt(x), None) # raiz quadrada "np.where(condition, value_if_true, value_if_false!"
e = lambda x: 10 * np.cbrt(x)  # raiz cúbica
f = lambda x: np.abs(x)  # módulo

# Funções trigonométricas e afins
g = lambda x: np.sin(x)  # seno
h = lambda x: np.cos(x)  # cosseno
i = lambda x: np.tan(x)  # tangente
j = lambda x: np.where((-1 < x) & (x < 1), np.arcsin(x), None)  # arco seno com seu domínio delimitado
k = lambda x: np.where((-1 < x) & (x < 1), np.arccos(x), None)  # arco cosseno com seu domínio delimitado
l = lambda x: np.arctan(x)  # arco

# Funções hiperbólicas e afins
m = lambda x: np.sinh(x)  # seno hiperbólico
n = lambda x: np.cosh(x)  # cosseno hiperbólico
o = lambda x: np.tanh(x)  # tangente hiperbólica
p = lambda x: np.where(x > 1, np.arcsinh(x), None) # arco seno hiperbólico com seu domínio delimitado
q = lambda x: np.where((-1 < x) & (1 < x), np.arccosh(x), None) # arco cosseno hiperbólico com seu domínio delimitado
r = lambda x: np.arctanh(x)  # arco tangente hiperbólico

# Funções exponenciais e logarítmicas
s = lambda x: np.exp(x)  # exponencial natural
t = lambda x: np.where(x > 0, np.log(x), None) # logaritmo natural com seu domínio delimitado
u = lambda x: np.where(x > 0, np.log10(x), None) # logaritmo base 10 com seu domínio delimitado
w = lambda x: np.exp2(x)  # exponencial base 2
````

## 2.1 Plotagem estática

Importação das bibliotecas

````python
# Importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

Declaração das funções e criação de intervalo e amostragem para a variável independente, "np.meshgrid(x,y)" e "Z = f(X,Y)" estão criando uma matriz de amostras que representa o plano xy e associando aos valores no eixo z.

````python
# Definindo a função de duas variáveis
def f(x, y):
    return x ** 2 - y ** 2

# Região a ser plotada
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y) # cria uma matriz de coordenadas (x, y)que cobre toda a área definida pelos intervalos de x e y.
Z = f(X, Y) # retorna uma matriz Z que contém os valores calculados de f para cada par de coordenadas (X[i, j], Y[i, j])
````

Plotando os gráficos com seus respectivos labels, cores e configurações dos eixos, legendas e do display de plotagem.

O trecho "ax = fig.add_subplot(111, projection='3d')" adiciona um subplot à figura fig. 111 significa que estamos criando um subplot com 1 linha, 1 coluna e este é o subplot número 1. Enquanto "projection='3d'" especifica que este subplot será em 3D, ou seja, um gráfico tridimensional.

Já em "ax.plot_surface(X, Y, Z, cmap='viridis')", ax é o objeto do subplot 3D criado anteriormente é usado para plotar uma superfície tridimensional no espaço 3D definido pelos arrays X, Y e Z. X, Y e Z são os dados de coordenadas e valores da superfície. "cmap='viridis'" especifica o esquema de cores a ser usado para a superfície.

````python
# Criando a figura e o gráfico 3D
fig = plt.figure(figsize=(10, 10)) # cria uma nova figura
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Nomeando os eixos e o gráfico
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Função de duas variáveis')

# Ajustando a visualização
ax.view_init(elev=50, azim=45)

# Mostrando o plot
plt.show()
````

````python
# Definindo a função de duas variáveis
def f(x, y):
    return x ** 2 - y ** 2

# Região a ser plotada
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
````

## 2.3 Plotagem com animação interativa

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import display, HTML
import ipywidgets as widgets

# Função para atualizar o plot a cada quadro da animação
def update(frame):
    ax.clear()
    ax.plot(x, np.sin(x + frame / 10.0))
    ax.set_ylim(-1.5, 1.5)
    ax.set_title(f'Frame {frame}')

# Criando a figura e o eixo
fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 100)

# Criando a animação
ani = FuncAnimation(fig, update, frames=100, interval=50)

# Criando um slider para controlar o frame
frame_slider = widgets.IntSlider(min=0, max=100, step=1, continuous_update=False)

# Função para atualizar a animação com base no valor do slider
def update_animation(change):
    ani.frame_seq = ani.new_frame_seq()  # Reinicia a sequência de frames
    ani.frame_seq = [ani.frame_seq[i] for i in range(len(ani.frame_seq)) if i <= change['new']]
    ani.event_source.start()

# Conectando a função de atualização ao slider
frame_slider.observe(update_animation, names='value')

# Exibindo o slider
display(frame_slider)

# Exibindo a animação
display(HTML(ani.to_jshtml()))
````
